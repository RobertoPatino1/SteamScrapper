```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df = pd.read_csv("../game_both_price.csv")

# Convertir los precios a nmeros (eliminando el smbolo $)
df['Precio Original'] = df['Precio Original'].replace('[\$,]', '', regex=True).astype(float)
df['Precio Final con Descuento'] = df['Precio Final con Descuento'].replace('[\$,]', '', regex=True).astype(float)

# Seleccionar las primeras 50 filas
df = df.head(25)

# Crear un grfico de barras
fig, ax = plt.subplots(figsize=(12, 8))

# Barra para Precio Original
ax.bar(df['Titulo'] + " - Original", df['Precio Original'], label='Precio Original')

# Barra para Precio Final con Descuento
ax.bar(df['Titulo'] + " - Descuento", df['Precio Final con Descuento'], label='Precio Final con Descuento')

# Configurar etiquetas y ttulo
ax.set_xlabel('Juego')
ax.set_ylabel('Precio ($)')
ax.set_title('Comparacin de Precios Originales y Finales con Descuento por Juego (Primeros 50 Juegos)')
ax.legend()

# Rotar las etiquetas del eje x para mayor claridad
plt.xticks(rotation=90, ha='right')

# Mostrar el grfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo CSV
df = pd.read_csv("../game_both_price.csv")

# Convertir los precios a nmeros (eliminando el smbolo $)
df['Precio Original'] = df['Precio Original'].replace('[\$,]', '', regex=True).astype(float)
df['Precio Final con Descuento'] = df['Precio Final con Descuento'].replace('[\$,]', '', regex=True).astype(float)

# Seleccionar las primeras 50 filas
df = df.head(50)

# Crear un grfico de barras agrupadas
fig, ax = plt.subplots(figsize=(15, 8))

# Definir el ancho de las barras y la posicin de los juegos
bar_width = 0.4
bar_positions = np.arange(len(df['Titulo']))

# Barra para Precio Original
ax.bar(bar_positions - bar_width/2, df['Precio Original'], width=bar_width, label='Precio Original')

# Barra para Precio Final con Descuento
ax.bar(bar_positions + bar_width/2, df['Precio Final con Descuento'], width=bar_width, label='Precio Final con Descuento')

# Configurar etiquetas y ttulo
ax.set_xlabel('Juego')
ax.set_ylabel('Precio ($)')
ax.set_title('Comparacin de Precios Originales y Finales con Descuento por Juego (Primeros 50 Juegos)')
ax.set_xticks(bar_positions)
ax.set_xticklabels(df['Titulo'], rotation=90, ha='right')
ax.legend()

# Mostrar el grfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df_languages = pd.read_csv("../games_by_language.csv")

# Crear un diagrama de burbujas con ejes cambiados
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(df_languages['Total de juegos disponibles'], df_languages['Idioma'], s=df_languages['Total de juegos disponibles']*0.001, alpha=0.5)

# Configuracin del ttulo
ax.set_title('Distribucin de Juegos por Idioma')

# Configuracin de los ejes
ax.set_xlabel('Total de juegos disponibles')
ax.set_ylabel('Idioma')

# Mostrar el diagrama de burbujas con ejes cambiados
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Leer el archivo CSV
df_vr_info = pd.read_csv("../vr_games_info.csv")

# Crear un grfico de donut
fig, ax = plt.subplots(figsize=(8, 8))

# Datos para el grfico de donut
sizes = [df_vr_info['Total de juegos con soporte VR'].values[0], df_vr_info['Total de juegos exclusivos para VR'].values[0]]
labels = ['Con soporte VR', 'Exclusivos para VR']
colors = ['#66b3ff', '#99ff99']

# Crear el grfico de donut
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=colors, wedgeprops=dict(width=0.4, edgecolor='w'))

# Configuracin del ttulo
ax.set_title('Distribucin de Juegos VR')

# Mostrar el grfico de donut
plt.show()
```

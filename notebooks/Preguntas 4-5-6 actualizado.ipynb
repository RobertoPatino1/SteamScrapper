```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Cargar los datos desde el archivo CSV
df = pd.read_csv("../datos_categoria_completa.csv")

# Crear un diccionario con el nombre de la categoría como clave y el promedio de reseñas como valor
data_dict = dict(zip(df['Nombre'], df['Promedio de Resenas']))

# Crear el objeto WordCloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(data_dict)

# Mostrar el gráfico de nube de palabras
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud - Promedio de Reseñas por Categoría')
plt.show()
```

```python


# Cargar los datos desde el archivo CSV
df = pd.read_csv("../datos_categoria_completa.csv")

# Crear un diagrama de cajas para "Promedio de Precios"
plt.figure(figsize=(12, 6))
box = plt.boxplot(df['Promedio de Precios'])
plt.title('Diagrama de Cajas - Promedio de Precios por Categoría')
plt.xlabel('Nombre')
plt.ylabel('Promedio de Precios')
plt.xticks([1], [""], rotation=45, ha='right')

# Añadir un resumen estadístico
summary = df['Promedio de Precios'].describe()

# Extraer el nombre de la categoría
min_index = df['Promedio de Precios'].idxmin()
max_index = df['Promedio de Precios'].idxmax()
min_category_name = df.loc[min_index]['Nombre']
max_category_name = df.loc[max_index]['Nombre']

plt.text(1.1, summary['min'], f'Min: {summary["min"]:.2f} ({min_category_name})', va='center', ha='left', color='green')
plt.text(1.1, summary['25%'], f'25%: {summary["25%"]:.2f}', va='center', ha='left', color='blue')
plt.text(1.1, summary['50%'], f'50%: {summary["50%"]:.2f}', va='center', ha='left', color='red')
plt.text(1.1, summary['75%'], f'75%: {summary["75%"]:.2f}', va='center', ha='left', color='blue')
plt.text(1.1, summary['max'], f'Max: {summary["max"]:.2f} ({max_category_name})', va='center', ha='left', color='green')

# Mostrar el diagrama de cajas
plt.show()
```

```python
# Cargar los datos desde el archivo CSV
df = pd.read_csv("../datos_categoria.csv")
df
categorias = df["Nombre"]
windows = df["windows"]
mac = df["mac"]
linux = df["linux"]

# Crear el gráfico de barras apiladas
plt.figure(figsize=(12, 6))
plt.bar(categorias, windows, label='Windows', color='blue')
plt.bar(categorias, mac, label='Mac', bottom=windows, color='orange')
plt.bar(categorias, linux, label='Linux', bottom=mac+windows, color='green')

plt.xlabel('Categorías de Juegos')
plt.ylabel('Cantidad')
plt.title('Distribución de Plataformas por Categoría de Juegos')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor legibilidad
plt.legend()
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

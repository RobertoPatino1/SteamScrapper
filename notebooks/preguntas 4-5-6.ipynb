```python
# Cargar los datos desde el archivo CSV
df = pd.read_csv("datos_categoria.csv")
df
```

```python
# Seleccionar las columnas "Nombre" y "Promedio de Reseñas"
categorias = df["Nombre"]
promedio_resenas = df["Promedio de Resenas"]

# Crear el gráfico de pastel
plt.figure(figsize=(10, 10))
plt.pie(promedio_resenas, labels=categorias, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.axis('equal')  # Hace que el gráfico de pastel sea circular
plt.title('Distribución del Promedio de Reseñas por Categoría de Juegos')

# Mostrar el gráfico
plt.show()
```

```python
categorias = df["Nombre"]
promedio_precios = df["Promedio de Precios"]

# Crear el gráfico de barras
plt.figure(figsize=(12, 6))
plt.bar(categorias, promedio_precios, color='skyblue')
plt.xlabel('Categorías de Juegos')
plt.ylabel('Promedio de Precios')
plt.title('Promedio de Precios por Categoría de Juegos')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor legibilidad
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
categorias = df["Nombre"]
windows = df["windows"]
mac = df["mac"]
linux = df["linux"]

# Crear el gráfico de barras apiladas
plt.figure(figsize=(12, 6))
plt.bar(categorias, windows, label='Windows', color='blue')
plt.bar(categorias, mac, label='Mac', bottom=windows, color='orange')
plt.bar(categorias, linux, label='Linux', bottom=mac+windows, color='green')

plt.xlabel('Categorías de Juegos')
plt.ylabel('Cantidad')
plt.title('Distribución de Plataformas por Categoría de Juegos')
plt.xticks(rotation=45, ha='right')  # Rotar las etiquetas del eje x para mejor legibilidad
plt.legend()
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
